# 1B. Function to get a book page

```python
def get_book_page(title):
    from bs4 import BeautifulSoup
    import requests
    try:
        request_url='https://www.goodreads.com/search.html?key=FUAhgnvFxs3aeW60paeX8A&q='+title
        r = requests.get(request_url)
        soup = BeautifulSoup(r.text, "xml")
        return soup
    except:
        print('No book found {}'.format(title))
```

# 1C. Function to parse book page data<br><i>(title, book_id, author_id, ratings_count, average_rating)</i>

```python
# will pass output of get_book_page function as argument in this function 
def get_data(book_page):
    
    all_data = []
    data_dict = {}
    
    try:
        title = str(book_page.find('title'))
        title = title[(title.index('>')+1):]
        title = title[0:(title.index('<'))]
        data_dict['title'] = title
    
    except:
        data_dict['title'] = 'NaN'
        print('No title found.')
    
    try:
        book_id = str(book_page.findAll('id', type="integer")[1])
        book_id = book_id[(book_id.index('>')+1):]
        book_id = book_id[0:(book_id.index('<'))]
        data_dict['book_id'] = book_id
        
    except:
        data_dict['book_id'] = 'NaN'
        print('No book_id found.')
        
        
    try:
        author_id = str(book_page.find('author'))
        author_id = author_id[(author_id.index('r">')+3):]
        author_id = author_id[0:(author_id.index('</'))]
        data_dict['author_id'] = author_id
    except:
        data_dict['author_id'] = 'NaN'
        print('No author_id found.')

    try:
        author = str(book_page.find('author'))
        author = author[(author.index('e>')+2):]
        author = author[0:(author.index('</'))]
        data_dict['author'] = author
    except:
        data_dict['author'] = 'NaN'
        print('No author found.')
    
    try:
        ratings_count = str(book_page.find('ratings_count'))
        ratings_count = ratings_count[(ratings_count.index('>')+1):]
        ratings_count = ratings_count[0:(ratings_count.index('<'))]
        data_dict['ratings_count'] = ratings_count
    except:
        data_dict['ratings_count'] = 'NaN'
        print('No ratings_count found.')
        
    try:
        average_rating = str(book_page.find('average_rating'))
        average_rating = average_rating[(average_rating.index('>')+1):]
        average_rating = average_rating[0:(average_rating.index('<'))]
        data_dict['average_rating'] = average_rating
    except:
        data_dict['average_rating'] = 'NaN'
        print('No average_rating found.')
    
    try:
        title_query = str(book_page.find('query'))
        title_query = title_query[(title_query.index('>')+1):]
        title_query = title_query[0:(title_query.index('<'))]
        data_dict['title_query'] = title_query
    except:
        data_dict['title_query'] = 'NaN'
        print('No title_query found.')
    
    all_data.append(data_dict)
    return all_data
```

```python
# function to request a webpage for a single author 
def get_author_page(auth_id):
    from bs4 import BeautifulSoup
    import requests
    try:
        request_url= "https://www.goodreads.com/author/show/"+auth_id+"?format=xml&key=FUAhgnvFxs3aeW60paeX8A&q"
        ra = requests.get(request_url)
        asoup = BeautifulSoup(ra.text, "xml")
        return asoup
    except:
        print('No author found {}'.format(auth_id))
```

```python
# function to obtain author ID, gender, and works_count for each author
def auth_data(author_page):
    
    all_ = []
    data_dict_ = {}
    
    author_id = str(author_page.find('author'))
    author_id = author_id[(author_id.index('d>')+2):]
    author_id = author_id[0:(author_id.index('</'))]
    data_dict_['author_id'] = author_id    
    
    if str(author_page.find('gender')) == "<gender/>":
        data_dict_['gender'] = 'None'
    else:
        gender = str(author_page.find('gender'))
        gender = gender[(gender.index('>')+1):]
        gender = gender[0:(gender.index('<'))]
        data_dict_['gender'] = gender

    works_count = str(author_page.find('works_count'))
    works_count = works_count[(works_count.index('>')+1):]
    works_count = works_count[0:(works_count.index('</'))]
    data_dict_['works_count'] = works_count

    all_.append(data_dict_)
    
    return all_
```

```python
def get_book_page(bk_id):
    from bs4 import BeautifulSoup
    import requests
    try:
        request_url= "https://www.goodreads.com/book/show/"+bk_id+"?key=FUAhgnvFxs3aeW60paeX8A&q"
        rb = requests.get(request_url)
        bsoup = BeautifulSoup(rb.text, "xml")
        return bsoup
    except:
        print('No book found {}'.format(bk_id))
```

```python
def book_data(book_page):
    
    all_books = []
    data_dict_b = {}
    
    bk_ids = str(book_page.find('id'))
    bk_ids = bk_ids[(bk_ids.index('d>')+2):]
    bk_ids = bk_ids[0:(bk_ids.index('</'))]
    data_dict_b['book_id'] = bk_ids    
    
    if str(book_page.find('num_pages')) == "<num_pages> </num_pages>":
        data_dict_b['num_pages'] = 'None'
    else:
        num_pages = str(book_page.find('num_pages'))
        num_pages = num_pages[(num_pages.index('>')+1):]
        num_pages = num_pages[0:(num_pages.index('<'))]  
        data_dict_b['num_pages'] = num_pages
    
    all_books.append(data_dict_b)
    
    return all_books
```
