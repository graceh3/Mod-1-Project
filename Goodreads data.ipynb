# Merging main dataframe containing Goodreads book data

```python
# Libraries needed for this notebook
import pandas as pd
import csv
```

```python
# importing the two main dataframes 
df1 = pd.read_csv('first_1k_main.csv')
df2 = pd.read_csv('remaining_main.csv')
```

```python
# merging two dataframes into one called main and dropping unnecessary column
main = pd.merge(df1,df2, how='outer')
main = main.drop('Unnamed: 0', axis = 1)
main.head()
```

```python
# upper-casing the contents in column title before running the comparison
series = main['title_query']
main['title'] = series.str.upper()
main.head()
```

```python
title_match = []
for index,row in main.iterrows():
    try:
        title_match.append(row['title'][0].startswith(row['title_query'][0]))
    except:
        title_match.append(False)
        print(row['title'],row['title_query'])
```

```python
# Number of true title matches
sum(title_match)
```

```python
# Number of false title mathces
len(title_match) - sum(title_match)
```

```python
# Adding new column to main dataframe containing results of the title_match
main["title_match"] = title_match
```

# Merging author dataframes containing author data

```python
dfa1 = pd.read_csv('auth_list1.csv')
dfa2 = pd.read_csv('auth_list2.csv')
```

```python
main_a = pd.merge(dfa1,dfa2, how='outer')
main_a = main_a.drop('Unnamed: 0', axis = 1)
main_a = main_a[['author_id','gender']].drop_duplicates()
main_a.sort_values(by='author_id')
main_a
```
