# Clean up the data to not be nested

We also decided to disrigard price

```python
#clean up data
#take out $date': {'$numberLong' etc
bs = []
for i in range(len(nyt)):
    parsed = {}
    parsed['amazon_product_url'] = nyt[i]['amazon_product_url']
    parsed['author'] = nyt[i]['author']
    parsed['title'] = nyt[i]['title']
    parsed['description'] = nyt[i]['description']
    parsed['bestsellers_date'] = nyt[i]['bestsellers_date']['$date']['$numberLong'] 
    parsed['published_date'] = nyt[i]['published_date']['$date']['$numberLong'] 
    parsed['publisher'] = nyt[i]['publisher']
    parsed['rank_on_list'] = nyt[i]['rank']['$numberInt']
    parsed['rank_last_week'] = nyt[i]['rank_last_week']['$numberInt']
    parsed['weeks_on_list'] = nyt[i]['weeks_on_list']['$numberInt']
    bs.append(parsed)


bs
```

# Put Data into a Pandas Data Frame

```python
nytbs = pd.DataFrame(bs)
```

```python
nytbs
```

# Fix the format of the book titles so that they could be put into a url to get data from Goodreads

```python
def fix_the_title(original_title):
    new = original_title.replace("'","%27").replace(' ','+').strip('?').strip('.').replace("’",'%27')
    return new
```

```python
nytbs['title_url'] = nytbs['title'].apply(lambda x: fix_the_title(x))
```

# Getting the max weeks on list an the min rank for a book

```python
weeks_on_list_df = pd.DataFrame(nytbs.groupby('title').weeks_on_list.max())
```

```python
weeks_on_list_df.head()

#reset the index
#merge to keep ALL rows
```

```python
weeks_on_list_df.rename(columns={'weeks_on_list': 'max_weeks_on_list'}, inplace=True)
weeks_on_list_df.head()
```

```python
print(nytbs[['title']].drop_duplicates(inplace=False).shape)
print(nytbs[['title']].shape)
```

```python
rank_df = pd.DataFrame(nytbs.groupby('title').rank_on_list.min())
rank_df.head()
```

```python
rank_df.rename(columns={'rank_on_list': 'min_rank'}, inplace=True)
rank_df.head()
```

```python
rank_df.reset_index().head()
```

```python
weeks_on_list_df.reset_index().head()
```

# Merge the 2 new Pandas DataFrames

```python
new = pd.merge(rank_df, weeks_on_list_df, how='left', on='title')
```

# Merge the new DataFrame with the original one

```python
df = pd.merge(nytbs, new, how='left', on='title')
```
