# Import Data

```python
import pandas as pd
```

```python
df = pd.read_csv('clean_data.csv')
```

```python
df
```

```python
df.sort_values(by=['ratings_count'])
```

# Data Visualization 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.pairplot(df);
```

# Average Rating vs Ratings count

```python
sns.jointplot(x="average_rating", y="ratings_count", data=df);
```

```python
sns.jointplot(x="average_rating", y="ratings_count", data=df, kind="reg");
```

# Average Rating

```python
sax = sns.countplot(x="average_rating", data=df)
```

```python
x = df.average_rating
plt.hist(x, bins =20)

plt.xlabel('Average Rating')
plt.ylabel('Frequency of Values')
plt.title('Histograms')
plt.show()
```

```python
x = df.average_rating
ax = sns.distplot(x)
ax.set(xlabel='Average Rating', ylabel='Frequency of Values')
plt.show()
```

```python
df.average_rating.describe() 
```

# Average rating vs Ratings Count part 2

```python
sns.regplot(x="average_rating", y="ratings_count", data=df);
```

```python
sns.lmplot(x="average_rating", y="ratings_count", data=df);
```

```python
sns.regplot(x='ratings_count', y="average_rating", data=df);
```

```python
sns.lmplot(x='ratings_count', y="average_rating", data=df);
```

```python
#best fit line taking out the outliers 
sns.lmplot(x="average_rating", y="ratings_count", data=df,
           robust=True, ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="average_rating", y="ratings_count", data=df,
           lowess=True);
```

# Comparing NYT to Goodreads

```python
sns.lmplot(x="average_rating", y="ratings_count", hue="max_weeks_on_list", data=df);
```

```python
sns.jointplot(x="average_rating", y="max_weeks_on_list", data=df, kind="reg");
```

```python
sns.jointplot(x="average_rating", y="min_rank", data=df, kind="reg");
```

```python
sns.jointplot(x="ratings_count", y="max_weeks_on_list", data=df, kind="reg");
```

```python
sns.jointplot(x="ratings_count", y="min_rank", data=df, kind="reg");
```

# Gender breakdown

```python
g = sns.FacetGrid(df, col="gender")
g.map(plt.scatter, "average_rating", "ratings_count", alpha=.7)
g.add_legend();
```

```python
g = sns.FacetGrid(df, col="gender")
g.map(plt.scatter, "max_weeks_on_list", "min_rank", alpha=.7)
g.add_legend();
```
