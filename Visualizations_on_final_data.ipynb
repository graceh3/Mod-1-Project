```python
df = pd.read_csv('clean_data.csv')
df.head()
```

```python
# cleaning up some of the columns that are not needed
df = df.drop(columns=['Unnamed: 0','Unnamed: 0.1','author_match','author_gr', 'author_id', 'book_id', 'title_query'])
```

```python
# clean NaNs in gender column
df = df.fillna({'gender':'None'})
df
```

```python
# clean NaNs in num_pages column
df = df.fillna({'num_pages':'None'})
df.head()
```
